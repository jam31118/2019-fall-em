```python
import numpy as np
from numpy import pi, sin, cos

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.patches import Circle
from matplotlib.colors import Normalize
from matplotlib import cm
```

```python
Rho, Phi = np.meshgrid(np.linspace(0,1,41), np.linspace(0,2*pi,51))
X, Y, Z = Rho * cos(Phi), Rho*sin(Phi), np.zeros_like(Rho) - 1.5

Colors = Rho*2

norm = Normalize(0, Colors.max(), clip=True)
C = cm.jet(norm(Colors))
```

```python
Z2, Phi2 = np.meshgrid(np.linspace(-1.5,0,21), np.linspace(0,2*pi,51))
X2, Y2 = 1.0*cos(Phi2), 1.0*sin(Phi2)

Colors2 = (1.5+Z2)

norm = Normalize(0, Colors2.max(), clip=True)
C2 = cm.jet(norm(Colors2))
```

```python
fig, ax = plt.subplots(subplot_kw={'projection':'3d'})

ax.plot_surface(X2,Y2,Z2, facecolors=C2[:-1,:-1], cmap='jet')
ax.plot_surface(X,Y,Z, facecolors=C, cmap='jet')
# circ = Circle((0,0),1)
# ax.add_artist(circ)

ax.view_init(elev=-30, azim=0)

ax.set_axis_off()
```
